RESTFUL API AND FLASK

1. What is a RESTful API?
  
   .    A REST API (Representational State Transfer Application Programming Interface) is a set of rules for designing web services that allows different software applications to communicate with each other over the internet using standard HTTP methods.

2. Explain the concept of API specification?

   . An API specification is a detailed, structured blueprint or contract that defines exactly how different software components should interact with an Application Programming Interface (API).  

3. What is Flask, and why is it popular for building APIs?

   . Flask is a lightweight and flexible Python web framework that provides the essentials for building web applications and APIs.

4. What is routing in Flask?

   .  Routing in Flask is the process of mapping URLs to specific Python functions, known as view functions, that handle the user request and return a response.

5. How do you create a simple Flask application?

   .  To create a simple Flask app, install Flask using pip, create an app.py file with Flask(_name_), define a route with @app.route('/'), return a message from the function, set FLASK_APP=app.py (or set FLASK_APP=app.py on Windows), then run flask run in your terminal to start the server and view "Hello, World!" in your browser.

6. What are HTTP methods used in RESTful APIs?

   .  REST APIs use standard HTTP methods (also called verbs) to perform actions on resources, primarily mapping to the standard CRUD (Create, Read, Update, Delete) operations.

7. What is the purpose of the @app.route() decorator in Flask?

   .  In Flask, the @app.route() decorator is used to map a specific URL path to a Python function (called a view function).

When a client (like a browser) requests that URL, Flask will execute the associated function and return its response to the client.

8. What is the difference between GET and POST HTTP methods?

   .  In HTTP, GET and POST are the two most commonly used request methods, each serving distinct purposes in client-server communication.

9. How do you handle errors in Flask APIs?

   .  Flask provides several mechanisms for handling errors, primarily through the @app.errorhandler decorator, the abort() function, and built-in Werkzeug exceptions.

10. How do you connect Flask to a SQL database?

     .   The most common and recommended way to connect a Flask application to a SQL database is by using the Flask-SQLAlchemy extension.

11. What is the role of Flask-SQLAlchemy?

     .  Flask-SQLAlchemy is a Flask extension that simplifies using SQLAlchemy, a powerful Python Object-Relational Mapper (ORM).

12. What are Flask blueprints, and how are they usefu?

     .  Flask blueprints are a way to organize a large Flask application into smaller, modular, and reusable components or "mini-applications".

13. What is the purpose of Flask's request object?

      .     The purpose of Flask's request object is to act as a container that encapsulates all the information sent by a client with an incoming HTTP request.

14. How do you create a RESTful API endpoint using FlasK?

     .   To create a RESTful API endpoint using Flask, you define routes using the @app.route decorator and specify the allowed HTTP methods

15. What is the purpose of Flask's jsonify() functioN?

     . The jsonify() function in Flask is used to serialize data into JSON format and return an HTTP response with the Content-Type header set to application/json.

 16. Explain Flask’s url_for() function?

     .    The url_for() function in Flask is used for URL building, generating a URL to a specific function dynamically.

17. How does Flask handle static files (CSS, JavaScript, etc.)?

     .   Flask handles static files such as CSS and JavaScript by serving them from a designated "static" folder within the application directory.

18. what is an API specification, and how does it help in building a Flask API?

      .  An API specification is a detailed documentation or contract that describes the endpoints, methods, parameters, and expected responses of an API.

19. What are HTTP status codes, and why are they important in a Flask API?

      .  HTTP status codes are standard three-digit numbers returned by a server to indicate the outcome of a client's request.

20. How do you handle POST requests in Flask?

     . POST requests are handled in Flask by specifying the methods=['POST'] argument in the @app.route() decorator for a view function.

21.  How would you secure a Flask API?

       . Various methods can be used to secure a Flask API.
Implement user authentication using extensions like Flask-Login or JWT (JSON Web Tokens).

22. What is the significance of the Flask-RESTful extension?

      . The Flask-RESTful extension simplifies the creation of REST APIs in Flask.
It provides abstractions and tools for building robust and scalable APIs.

23.What is the role of Flask’s session object?

     .      The session object in Flask is used to store data specific to a single user across multiple requests.
It allows the application to remember stateful information about the user.
